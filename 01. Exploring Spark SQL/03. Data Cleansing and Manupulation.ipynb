# Prerequisites

### Load employee.csv into DataFrame


```python

df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
                    sep="|",
                    inferSchema=True,
                    header=True,
                    quote="'")
display(df)
```

# Select Operations


```python
df.select("id","name").show(3)
```

```python
from pyspark.sql.functions import col
df.select(col("id"), col("col_skills").alias("skills")).show(3, truncate=False)
```

# Filter Operation


```python
df.filter((col("desig")=="Developer") & (col("company")=="Infosys")).select("id","name","desig","company").display()
```

```python
df.filter((col("desig")=="Devloper") | (col("gen")=="M")).select("id","name","desig","gen").display()
```

```python
df.filter((col("exp").between(1,6)) & (col('gen')=="M")).display()
```

# Grouping, Aggretion and Sorting


```python
df.groupBy(col("gen")).count().sort(col("count")).display()
```

```python
df.groupBy(col("gen")).count().sort(col("count").desc()).display()
```

```python
from pyspark.sql.functions import avg, sum, min, max, count
df.groupBy(col("desig")).agg(min("exp").alias("Minimum Experience"),max("exp").alias("Maximum Experience"),avg("exp").alias("Average Experence")).display()
```

```python
from pyspark.sql.functions import lit
df.withColumn("Is_employeed", lit("Yes")).display()
```

```python
from pyspark.sql.functions import when

df.withColumn(
    "Category",
    when(col("exp") > 9, "Senior")
    .when(col("exp") > 4, "Mid level")
    .when(col("exp") >= 0, "Junior")
    .otherwise("Invalid Experience")
).display()
```

# Droping Missing Records

```python
from pyspark.sql.functions import col
df.filter((col("id").isNotNull()) & (col("name").isNotNull())).display()
```

```python
df.fillna(0).fillna("Null in source").display()
```

```python
df.na.fill({
    "id":-1,
    "name":"Null in source",
    "exp":0,
    "gen":"NA"
}).display()
```
